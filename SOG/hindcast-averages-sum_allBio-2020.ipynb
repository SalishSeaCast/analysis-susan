```python
t2020_jun = xr.open_dataset('/data/sallen/results/MEOPAR/averages/hindcast.201905/SalishSea_1m_202006_202006_prod_T.nc')
t2020_jul = xr.open_dataset('/data/sallen/results/MEOPAR/averages/hindcast.201905/SalishSea_1m_202007_202007_prod_T.nc')
t2020_aug = xr.open_dataset('/data/sallen/results/MEOPAR/averages/hindcast.201905/SalishSea_1m_202008_202008_prod_T.nc')
```

```python
t2020_totalphyto = (t2020_jun['PPDIAT'][0] + t2020_jun['PPPHY'][0] 
                    + t2020_jun['PPMRUB'][0] +
                  t2020_jul['PPDIAT'][0] + t2020_jul['PPPHY'][0] 
                    + t2020_jul['PPMRUB'][0] +
                  t2020_aug['PPDIAT'][0] + t2020_aug['PPPHY'][0] 
                    + t2020_aug['PPMRUB'][0])/3.
```

```python
t2020_flag = (t2020_jun['PPPHY'][0] + t2020_jul['PPPHY'][0] 
                    + t2020_aug['PPPHY'][0])/3.
```

```python
dia2_t2020_jun = xr.open_dataset('/data/sallen/results/MEOPAR/averages/hindcast.201905/SalishSea_1m_202006_202006_dia2_T.nc')
dia2_t2020_jul = xr.open_dataset('/data/sallen/results/MEOPAR/averages/hindcast.201905/SalishSea_1m_202007_202007_dia2_T.nc')
dia2_t2020_aug = xr.open_dataset('/data/sallen/results/MEOPAR/averages/hindcast.201905/SalishSea_1m_202008_202008_dia2_T.nc')
```

```python
t2020_totalgraz = (dia2_t2020_jun['GRMESZDIAT'][0] + dia2_t2020_jun['GRMESZPHY'][0] 
                    + dia2_t2020_jun['GRMESZMRUB'][0] +
                  dia2_t2020_jun['GRMESZPON'][0] + dia2_t2020_jun['GRMESZMICZ'][0] 
                  + dia2_t2020_jul['GRMESZDIAT'][0] + dia2_t2020_jul['GRMESZPHY'][0] 
                    + dia2_t2020_jul['GRMESZMRUB'][0] +
                  dia2_t2020_jul['GRMESZPON'][0] + dia2_t2020_jul['GRMESZMICZ'][0]
                  + dia2_t2020_aug['GRMESZDIAT'][0] + dia2_t2020_aug['GRMESZPHY'][0] 
                    + dia2_t2020_aug['GRMESZMRUB'][0] +
                  dia2_t2020_aug['GRMESZPON'][0] + dia2_t2020_aug['GRMESZMICZ'][0] 
                    )/3.
```

```python
print (mymesh.gdept_1d[0, 20:23])
lz = 21
thickness = e3t[:lz+1].sum()
```

```python
for i in range(lz):
    mean_totalphyto[i] = mean_totalphyto[i] * e3t[i] 
    mean_flag[i] = mean_flag[i] * e3t[i] 
    mean_totalgraz[i] = mean_totalgraz[i] * e3t[i]     
mean_totalphyto = mean_totalphyto[:lz].sum(axis=0)
mean_phyto_ratio = mean_flag[:lz].sum(axis=0)/mean_totalphyto
mean_totalgraz = mean_totalgraz[:lz].sum(axis=0)
print (mean_totalphyto.shape)
```

```python
for i in range(lz):
    t2020_totalphyto[i] = t2020_totalphyto[i] * e3t[i] 
    t2020_flag[i] = t2020_flag[i] * e3t[i] 
    t2020_totalgraz[i] = t2020_totalgraz[i] * e3t[i] 
t2020_totalphyto = t2020_totalphyto[:lz].sum(axis=0)
t2020_phyto_ratio = t2020_flag[:lz].sum(axis=0)/t2020_totalphyto
t2020_totalgraz = t2020_totalgraz[:lz].sum(axis=0)
print (t2020_totalphyto.shape)
```

```python
imin, imax = 350, 750
jmin, jmax = 100, 350
physcale = 86400 # per day

cmap = cmocean.cm.algae
cmap.set_bad('burlywood')
cmap2 = cmocean.cm.balance
cmap2.set_bad('burlywood')

fig, axs = plt.subplots(1, 3,figsize = (7.5*3, 9.5*1))

#### Primary Production ####
labelstring ='Primary Productivity ($\mu$mol N m$^{-2}$ day$^{-1}$)'
vmax = 30
vmin = 5
s = axs[0].pcolormesh(physcale*np.ma.masked_array(mean_totalphyto[imin:imax, jmin:jmax],
                                           mask = 1 - mymesh.variables['tmask'][0,lz,imin:imax, jmin:jmax]), 
                        cmap = cmap, vmin=vmin, vmax=vmax)
colorbar = fig.colorbar(s, ax=axs[0])
colorbar.set_label(labelstring)
viz_tools.set_aspect(axs[0])
axs[0].set_title('Summer Mean 2007-2020');

s = axs[1].pcolormesh(physcale*np.ma.masked_array(t2020_totalphyto[imin:imax, jmin:jmax],
                                           mask = 1 - mymesh.variables['tmask'][0,lz,imin:imax, jmin:jmax]), 
                        cmap = cmap, vmin=vmin, vmax=vmax)
colorbar = fig.colorbar(s, ax=axs[1])
colorbar.set_label(labelstring)
viz_tools.set_aspect(axs[1])
axs[1].set_title('Summer 2020')

vmax = 25
vmin = -vmax
s = axs[2].pcolormesh(100*np.ma.masked_array(t2020_totalphyto[imin:imax, jmin:jmax] 
                                           / mean_totalphyto[imin:imax, jmin:jmax] - np.ones([imax-imin, jmax-jmin]), 
                                           mask = 1 - mymesh.variables['tmask'][0,lz,imin:imax, jmin:jmax]), 
                        cmap = cmap2, vmin=vmin, vmax=vmax)
colorbar = fig.colorbar(s, ax=axs[2])
colorbar.set_label('Percent of Mean')
viz_tools.set_aspect(axs[2])
axs[2].set_title('Anomaly 2020')

for ax in [axs[0], axs[1], axs[2]
           ]:
    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)
    ax.text(5, 105, 'Vancouver Island', fontsize=13)
    ax.text(150, 205, 'Sunshine Coast', fontsize=12);
    
fig.savefig('SummerProductivity2020.png')
```
