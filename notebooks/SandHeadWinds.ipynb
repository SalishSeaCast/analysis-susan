```python
import datetime as dt
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sandheads = pd.read_csv('SH_total_31Dec2020.dat', names=['Day', 'Month', 'Year', 'Hour', 'Along', 'Across'], sep='\s+')
```

```python
sandheads['datetime'] = pd.to_datetime(sandheads.drop(columns=['Along', 'Across']))
```

```python
sandheads = sandheads.set_index('datetime')
```

```python
sandheads['windspeed3'] = (sandheads['Along']**2 + sandheads['Across'])**(3/2) 
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 4))
sandheads['windspeed3'].plot(ax=ax);
```

```python
bymonth = sandheads['windspeed3'].resample('3M').mean()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 4))
bymonth.plot();
ax.grid()
ax.set_ylim(0, 200);
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 4))
bymonth.plot();
ax.grid()
ax.set_ylim(0, 200)
ax.set_xlim(dt.datetime(2007, 1, 1), dt.datetime(2020, 1, 1));
```
