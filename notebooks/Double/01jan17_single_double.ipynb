```python
iz = 0
ih = 12
imin, imax = 0, 898
jmin, jmax = 0, 398
fig, axs = plt.subplots(1, 3, figsize=(18, 10))
single.vosaline[ih, iz, imin:imax, jmin:jmax].plot(ax=axs[0], cmap=cm.haline, vmin=10, vmax=28)
double.vosaline[ih, iz, imin*2:imax*2, jmin*2:jmax*2].plot(ax=axs[1], cmap=cm.haline, vmin=10, vmax=28)
((double.vosaline[ih, iz, imin*2:imax*2:2, jmin*2:jmax*2:2] - 
  single.vosaline[ih, iz, imin:imax, jmin:jmax]) * stmask[0, iz, imin:imax, jmin:jmax]).plot(ax=axs[2], vmax=2, vmin=-2, cmap=cm.balance);
axs[0].plot(155, 670, 'ro')
```

```python
ih = 12
ii = 426
jmin, jmax = 278, 314
fig, axs = plt.subplots(1, 3, figsize=(20, 7))
single.vosaline[ih, :, ii, jmin:jmax].plot(ax=axs[0], cmap=cm.haline, yincrease=False)
double.vosaline[ih, :, ii*2, jmin*2:jmax*2].plot(ax=axs[1], cmap=cm.haline, yincrease=False)
axs[1].plot(520, 695, 'ro')
mydouble = np.array(double.vosaline[ih, ::2, ii*2, jmin*2:jmax*2:2])
mysingle = np.array(single.vosaline[ih, :, ii, jmin:jmax])
mymask = np.array(stmask[0, :, ii, jmin:jmax])
colours = axs[2].pcolormesh((mydouble - mysingle) * mymask, vmin=-10, vmax=10, cmap=cm.balance)
fig.colorbar(colours, ax=axs[2])
for ax in axs:
    ax.set_ylim(20,0)
```
