```python
import matplotlib.pyplot as plt
import xarray as xr

%matplotlib inline
```

```python
solar = xr.open_dataset('solar.nc')
prev = xr.open_dataset('/data/dlatorne/tmp-rpn-to-gem-lam/gemlam_y2014m11d04_017.nc')
orig = xr.open_dataset('/data/dlatorne/tmp-rpn-to-gem-lam/gemlam_y2014m11d04_018.nc')
final = xr.open_dataset('gemlam_y2014m11d04_018.nc')
precip = xr.open_dataset('precip.nc')
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
prev['solar'].plot(ax=axs[0, 0])
orig['solar'].plot(ax=axs[0, 1])
(0.5*(prev['solar'][0]+orig['solar'][0])).plot(ax=axs[1, 0])
solar['solar'].plot(ax=axs[1, 1])
final['solar'].plot(ax=axs[1, 2]);
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
prev['precip'].plot(ax=axs[0, 0])
orig['precip'].plot(ax=axs[0, 1])
((orig['precip'][0] - prev['precip'][0])).plot(ax=axs[1, 0])
precip['precip'].plot(ax=axs[1, 1]);
(precip['precip'][0]-(orig['precip'][0] - prev['precip'][0])).plot(ax=axs[1, 2])
```
