```python
lost = []
lost.append(pd.read_csv('/data/sallen/results/Ariane/FullSouth/lost_particles', sep=' ', engine='python', header=None))
lost.append(pd.read_csv('/data/sallen/results/Ariane/BackNorth/lost_particles', sep=' ', engine='python', header=None))
lost.append(pd.read_csv('/data/sallen/results/Ariane/InGIslands/lost_particles', sep=' ', engine='python', header=None))
lost.append(pd.read_csv('/data/sallen/results/Ariane/FullNorth/lost_particles', sep=' ', engine='python', header=None))
lost.append(pd.read_csv('/data/sallen/results/Ariane/BackSouth/lost_particles', sep=' ', engine='python', header=None))
lost.append(pd.read_csv('/data/sallen/results/Ariane/SouthGIslands/lost_particles', sep=' ', engine='python', header=None))
lost[5]
```

```python
total = []
total.append(pd.read_csv('/data/sallen/results/Ariane/FullSouth/total_particles', sep=' ', engine='python', header=None))
total.append(pd.read_csv('/data/sallen/results/Ariane/BackNorth/total_particles', sep=' ', engine='python', header=None))
total.append(pd.read_csv('/data/sallen/results/Ariane/InGIslands/total_particles', sep=' ', engine='python', header=None))
total.append(pd.read_csv('/data/sallen/results/Ariane/FullNorth/total_particles', sep=' ', engine='python', header=None))
total.append(pd.read_csv('/data/sallen/results/Ariane/BackSouth/total_particles', sep=' ', engine='python', header=None))
total.append(pd.read_csv('/data/sallen/results/Ariane/SouthGIslands/total_particles', sep=' ', engine='python', header=None))
total[5]
```
