```python
import matplotlib.pyplot as plt
import xarray as xr

%matplotlib inline
```

```python
iY = 344; iX = 276; iZ = 10 
```

```python
ds1 = xr.open_dataset('/Users/sallen/Documents/E471/eworksheets/Miniproject3/SalishSea_1h_20160801_20160801_grid_U.nc')
velocity = ds1['vozocrtx'].isel(y=iY, x=iX).sel(depthu=iZ, method='nearest')
```

```python
velocity.plot()
```

```python
velocity_2016 = velocity.copy(deep=True)
velocity_2016.plot()
```

```python
velocity = 0.
ds1.close()
velocity_2016.plot()
```

```python
ds2 = xr.open_dataset('/Users/sallen/Documents/E471/eworksheets/Miniproject3/SalishSea_1h_20160802_20160802_grid_U.nc')
velocity = ds2['vozocrtx'].isel(y=iY, x=iX).sel(depthu=iZ, method='nearest')
velocity.plot();
```
