Zero the Ciliates in the Restart File

```python
import numpy as np
import xarray as xr

%matplotlib inline
```

```python
tracerrestart = '/results2/SalishSea/nowcast-green.201905/31oct14/SalishSea_03021840_restart_trc.nc'
```

```python
df = xr.open_dataset(tracerrestart)
```

```python
df.keys()
mesorub_keys = ['rnf_pis_MYRI_b', 'sbc_MYRI_b', 'TRNMYRI', 'TRBMYRI']
```

```python
df[mesorub_keys[2]][0, 0].plot();
```

```python
for key in mesorub_keys:
    df[key] = 0.*df[key]
```

```python
df[mesorub_keys[2]][0, 0].plot();
```

```python
df
```

```python
df.attrs['Manipulation'] = 'All Mesodinium rubrum fields set to zero'
netcdf_file = 'SalishSea_ZeroCiliates_03021840_restart_trc.nc'
df.attrs['file_name'] = netcdf_file
df.attrs
```

```python
encoding = {var: {'zlib': True} for var in df.data_vars}

df.to_netcdf(netcdf_file, unlimited_dims=('t'),
        encoding=encoding,)
```

```python
ds = xr.Dataset(
    {"zero": (["time_counter", "y", "x"], np.zeros((1, 898, 398)))},
    coords={
        "time_counter": np.array([0]),
        "y": np.arange(898),
        "x": np.arange(398),
       },
    attrs = {'summary': 'Flat field of zeros'}

)
```

```python
ds
```

```python
encoding = {'zero': {'zlib': True} }

ds.to_netcdf('all_zeros.nc', unlimited_dims=('time_counter'),
        encoding=encoding,)
```
