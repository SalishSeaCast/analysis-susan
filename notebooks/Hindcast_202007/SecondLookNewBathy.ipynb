# Plume Location June 2017

```python
sal_Gp1 = xr.open_dataset('/home/sallen/202007/202007G-p1/SalishSea_1m_201706_201706_grid_T.nc')
sal_F = xr.open_dataset('/home/sallen/202007/202007F/SalishSea_1m_201706_201706_grid_T.nc')
```

```python
kk = 0
vvmax = 10
vmin, vmax = 8, 26
fig, axs = plt.subplots(1, 3, figsize=(15, 8))
colours = axs[0].pcolormesh(np.ma.masked_array(sal_F.vosaline[0, kk], mask=1-tmask1[kk]), cmap=cm.haline, vmin=vmin, vmax=vmax)
fig.colorbar(colours, ax=axs[0]);
colours = axs[1].pcolormesh(np.ma.masked_array(sal_Gp1.vosaline[0, kk], mask=1-tmask2[kk]), cmap=cm.haline, vmin=vmin,vmax=vmax)
fig.colorbar(colours, ax=axs[1]);
colours = axs[2].pcolormesh(np.ma.masked_array(sal_Gp1.vosaline[0, kk] - sal_F.vosaline[0, kk], mask=1-tmask[kk]), cmap=cm.balance, 
                            vmax=vvmax, vmin=-vvmax);
fig.colorbar(colours, ax=axs[2]);
for ax in axs:
    viz_tools.set_aspect(ax)
axs[0].set_title('Surface Salinity, Run F: Old Bathy')
axs[1].set_title('Surface Salinity, Run Gp1: New Bathy')
axs[2].set_title('New Bathy - Old Bathy');
```

```python
kk = 10
vvmax = 10
vmin, vmax = 18*kk, 28*kk
fig, axs = plt.subplots(1, 3, figsize=(15, 8))
colours = axs[0].pcolormesh(np.ma.masked_array(sal_F.vosaline[0, 0:kk].sum(axis=0), mask=1-tmask1[kk]), cmap=cm.haline, vmin=vmin, vmax=vmax)
fig.colorbar(colours, ax=axs[0]);
colours = axs[1].pcolormesh(np.ma.masked_array(sal_Gp1.vosaline[0, 0:kk].sum(axis=0), mask=1-tmask2[kk]), cmap=cm.haline, vmin=vmin, vmax=vmax)
fig.colorbar(colours, ax=axs[1]);
colours = axs[2].pcolormesh(np.ma.masked_array(sal_Gp1.vosaline[0, 0:kk].sum(axis=0) - sal_F.vosaline[0, 0:kk].sum(axis=0), mask=1-tmask[kk]), cmap=cm.balance, 
                            vmax=vvmax, vmin=-vvmax);
fig.colorbar(colours, ax=axs[2]);
for ax in axs:
    viz_tools.set_aspect(ax)
axs[0].set_title('Top 10 m integrated Salinity, Run F: Old Bathy')
axs[1].set_title('op 10 m integratedSalinity, Run Gp1: New Bathy')
axs[2].set_title('New Bathy - Old Bathy');
```
