```python
import cmocean.cm as cm
import matplotlib.pyplot as plt
import netCDF4 as nc
import numpy as np
import xarray as xr

from salishsea_tools import viz_tools

%matplotlib inline
```

```python
mesh = nc.Dataset('/home/sallen/MEOPAR/grid/mesh_mask201702.nc')
tmask = mesh['tmask']
```

```python
new = '/home/sallen/202007/202007C-p3/SalishSea_1h_20150101_20150630_grid_T_20150105-20150105.nc'
old = '/home/sallen/202007/202007C-p2/SalishSea_1h_20150101_20151231_grid_T_20150105-20150105.nc'
```

```python
rivers = xr.open_dataset(new)
climate = xr.open_dataset(old)
```

```python
vmin, vmax= 0, 30
fig, axs = plt.subplots(1, 3, figsize=(15, 7))
orig = np.ma.array(climate.vosaline[23, 0], mask=1-tmask[0, 0])
colours = axs[0].pcolormesh(orig, cmap=cm.haline, vmax=vmax, vmin=vmin)
axs[0].set_title("Fraser Daily, Climate Else")
fig.colorbar(colours, ax=axs[0])
daily = np.ma.array(rivers.vosaline[23, 0], mask=1-tmask[0, 0])
colours = axs[1].pcolormesh(daily, cmap=cm.haline, vmax=vmax, vmin=vmin)
axs[1].set_title("All Daily")
fig.colorbar(colours, ax=axs[1])
colours = axs[2].pcolormesh(daily - orig, cmap=cm.balance, vmin=-5, vmax=5)
axs[2].set_title("Difference")
fig.colorbar(colours, ax=axs[2])
for ax in axs:
    viz_tools.set_aspect(ax)
```

```python
def do_plots(year, month, day, dl=0):
    new = f'/home/sallen/202007/202007C-p3/SalishSea_1h_20150101_20150630_grid_T_{year}{month:02d}{day:02d}-{year}{month:02d}{day:02d}.nc'
    old = f'/home/sallen/202007/202007C-p2/SalishSea_1h_20150101_20151231_grid_T_{year}{month:02d}{day:02d}-{year}{month:02d}{day:02d}.nc'
    rivers = xr.open_dataset(new)
    climate = xr.open_dataset(old)
    
    params = {'full': {'vmin': 0,
                      'vmax': 30,
                      'imin': 0,
                      'imax': -1,
                      'jmin': 0,
                      'jmax': -1}, 
             'plume': {'vmin': 0,
                      'vmax': 30,
                      'imin': 350,
                      'imax': 550,
                      'jmin': 250,
                      'jmax': 350},
             'jdf': {'vmin': 10,
                      'vmax': 30,
                      'imin': 300,
                      'imax': 500,
                      'jmin': 0,
                      'jmax': 100},
             'whitbey': {'vmin': 10,
                      'vmax': 30,
                      'imin': 100,
                      'imax': 300,
                      'jmin': 250,
                      'jmax': 350}
             }
    
    
    orig = np.ma.array(climate.vosaline[23, dl], mask=1-tmask[0, dl])
    daily = np.ma.array(rivers.vosaline[23, dl], mask=1-tmask[0, dl])
    
    fig, axs = plt.subplots(4, 3, figsize=(15, 28))
    axs[0, 0].set_title("Fraser Daily, Climate Else")
    axs[0, 1].set_title("All Daily")
    axs[0, 2].set_title("Difference")
    
    for row, key in enumerate(params.keys()):
        imin, imax, jmin, jmax = params[key]['imin'], params[key]['imax'], params[key]['jmin'], params[key]['jmax']
        vmin, vmax = params[key]['vmin'], params[key]['vmax']

        colours = axs[row, 0].pcolormesh(orig[imin:imax, jmin:jmax], cmap=cm.haline, vmax=vmax, vmin=vmin)
        fig.colorbar(colours, ax=axs[row, 0])

        colours = axs[row, 1].pcolormesh(daily[imin:imax, jmin:jmax], cmap=cm.haline, vmax=vmax, vmin=vmin)
        fig.colorbar(colours, ax=axs[row, 1])

        colours = axs[row, 2].pcolormesh(daily[imin:imax, jmin:jmax] - orig[imin:imax, jmin:jmax], cmap=cm.balance, vmin=-5, vmax=5)
        fig.colorbar(colours, ax=axs[row,2])
        for ax in axs[row]:
            viz_tools.set_aspect(ax)
            
    rivers.close()
    climate.close()
```

```python
do_plots(2015, 1, 31)
```

```python
do_plots(2015, 2, 7)
```

```python
do_plots(2015, 2, 16, dl=4)
```

```python
do_plots(2015, 3, 9, dl=0)
```
