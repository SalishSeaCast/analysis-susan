```python
import matplotlib.pyplot as plt
import netCDF4 as nc
import numpy as np

%matplotlib inline
```

```python
LOfile = '/results/forcing/LiveOcean/modified/bio/single_bio_LO_y2017m08d09.nc'
LOfile_prev = '/results/forcing/LiveOcean/modified/bio/single_bio_LO_y2017m08d07.nc'
LO_ds = nc.Dataset(LOfile)
LO_ps = nc.Dataset(LOfile_prev)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
nitrate = LO_ds['NO3'][:, :, :]
nitrate_prev = LO_ps['NO3'][:, :, :]
print(nitrate.shape)
mesh = ax[0].pcolormesh(nitrate_prev[0, :, 0], cmap='viridis')
fig.colorbar(mesh, ax = ax[0])
mesh = ax[1].pcolormesh(nitrate[0, :, 0], cmap='viridis')
fig.colorbar(mesh, ax = ax[1])
mesh = ax[2].pcolormesh(nitrate[0, :, 0] - nitrate_prev[0, :, 0], cmap='bwr', vmax=1, vmin=-1)
fig.colorbar(mesh, ax = ax[2])
```
