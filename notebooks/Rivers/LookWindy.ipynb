```python
import cmocean.cm as cm
import netCDF4 as nc
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
days = ['y2016m12d09', 'y2016m12d18']
imin = 879
jmin = 35
jmax = 60
```

```python
fig, axs = plt.subplots(1,3, figsize=(15, 5))
for ax, day in zip(axs[0:2],days):
    riverfile = nc.Dataset('/data/sallen/MEOPAR/forcing/RLonFraCElse_'+day+'.nc')
    flow = riverfile.variables['rorunoff'][0, imin:-1, jmin:jmax]
    mesh = ax.pcolormesh(flow, cmap=cm.haline)
    riverfile.close()
    fig.colorbar(mesh, ax=ax)
    ax.set_title(day)
riverfile = nc.Dataset('/ocean/sallen/allen/research/MEOPAR/NEMO-forcing/rivers/rivers_month_extended.nc')
flow = riverfile.variables['rorunoff'][9, imin:-1, jmin:jmax]
mesh = axs[2].pcolormesh(flow, cmap=cm.haline)
riverfile.close()
fig.colorbar(mesh, ax=axs[2])
axs[2].set_title('New, mid October')
```

```python
fig, axs = plt.subplots(1,3, figsize=(15, 5))
riverfile = nc.Dataset('/data/sallen/MEOPAR/forcing/old_'+days[0]+'.nc')
flow1 = riverfile.variables['rorunoff'][0, imin:-1, jmin:jmax]
riverfile.close()
riverfile = nc.Dataset('/data/sallen/MEOPAR/forcing/RLonFraCElse_'+days[0]+'.nc')
flow2 = riverfile.variables['rorunoff'][0, imin:-1, jmin:jmax]
riverfile.close()

for ax, flow in zip(axs[0:2], [flow1, flow2]):
    mesh = ax.pcolormesh(flow, cmap=cm.haline)
    fig.colorbar(mesh, ax=ax)
flow = flow2-flow1
mesh = axs[2].pcolormesh(flow, cmap=cm.haline)
fig.colorbar(mesh, ax=axs[2])
```
