```python
import cmocean.cm as cm
import netCDF4 as nc
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
days = ['y2016m09d15', 'y2016m11d15']
imin = 879
jmin = 35
jmax = 60
```

```python
fig, axs = plt.subplots(1,3, figsize=(15, 5))
for ax, day in zip(axs[0:2],days):
    riverfile = nc.Dataset('/results/forcing/rivers/RLonFraCElse_'+day+'.nc')
    flow = riverfile.variables['rorunoff'][0, imin:-1, jmin:jmax]
    mesh = ax.pcolormesh(flow, cmap=cm.haline)
    riverfile.close()
    fig.colorbar(mesh, ax=ax)
    ax.set_title(day)
riverfile = nc.Dataset('/ocean/sallen/allen/research/MEOPAR/Tools/I_ForcingFiles/Rivers/rivers_month_extended.nc')
flow = riverfile.variables['rorunoff']
mesh = ax.pcolormesh(flow, cmap=cm.haline)
riverfile.close()
fig.colorbar(mesh, ax=ax)
ax.set_title('New, mid October')
```
