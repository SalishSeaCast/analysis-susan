```python
import matplotlib.pyplot as plt
import xarray as xr

%matplotlib inline
```

```python
df = xr.open_dataset('/home/sallen/MEOPAR/average_2017_grid_T.nc')
ssh = df.sossheig
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 9))
ssh.plot(vmax=0.5, vmin=-0.5, cmap='bwr')
```
