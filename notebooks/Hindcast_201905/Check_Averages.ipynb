```python
import cmocean.cm as cm
import matplotlib.pyplot as plt
import xarray as xr

%matplotlib inline
```

```python
# Choose area, Rosario
```

```python
imin = 300; imax = 400; jmin = 250; jmax = 350
```

```python
hourly = xr.open_dataset('/results2/SalishSea/nowcast-green.201905/09jun21/SalishSea_1h_20210609_20210609_grid_T.nc')
daily = xr.open_dataset('/results2/SalishSea/nowcast-green.201905/09jun21/SalishSea_1d_20210609_20210609_grid_T.nc')
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 6))
hourly.votemper[:, 0, imin:imax, jmin:jmax].mean(axis=0).plot(ax=axs[0], cmap=cm.thermal, vmin=8, vmax=18)
daily.votemper[0, 0, imin:imax, jmin:jmax].plot(ax=axs[1], cmap=cm.thermal, vmin=8, vmax=18)
(daily.votemper[0, 0, imin:imax, jmin:jmax]-hourly.votemper[:, 0, imin:imax, jmin:jmax].mean(axis=0)).plot(ax=axs[2], cmap=cm.curl);
```
