```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr

%matplotlib inline
```

```python
orig = xr.open_dataset('/results/forcing/LiveOcean/boundary_conditions/LiveOcean_v201905_y2020m08d12.nc')
new = xr.open_dataset('/results/forcing/LiveOcean/boundary_conditions/LiveOcean_v201905_test_y2020m08d12.nc')
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
orig.vosaline[0, :, 0].plot(ax=axs[0])
new.vosaline[0, :, 0].plot(ax=axs[1])
(new.vosaline - orig.vosaline)[0, :, 0].plot(ax=axs[2], vmax=0.1, vmin=-0.1, cmap='bwr');
print(np.array(np.abs(new.vosaline - orig.vosaline)).max())
```

```python
key = 'OXY'
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
orig[key][0, :, 0].plot(ax=axs[0])
new[key][0, :, 0].plot(ax=axs[1])
(new[key] - orig[key])[0, :, 0].plot(ax=axs[2], vmax=5, vmin=-5, cmap='bwr');
print(np.array(np.abs(new[key] - orig[key])).max())
```
