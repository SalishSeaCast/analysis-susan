Find the index of the first z-grid point with salinity > 28 and exclude all regions with no salinity > 28

```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr

%matplotlib inline
```

```python
tmask = xr.open_dataset('/home/sallen/MEOPAR/grid/mesh_mask201702.nc')['tmask'][0]
```

```python
with xr.open_dataset('/results2/SalishSea/hindcast.201812_annex/01jun19/SalishSea_1d_20190601_20190601_grid_T.nc') as ds:
    salinity = ds['vosaline'][0]
    depthindex = np.argmax(salinity>28, axis=0)
    maxsal = np.amax(salinity, axis=0)>28
```

```python
plt.pcolormesh(np.ma.array(depthindex, mask=(1 - (tmask[0] * maxsal))));
plt.colorbar();
```
