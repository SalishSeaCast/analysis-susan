```python
# a year!
```

```python
mypath = Path('/results/forcing/atmospheric/GEM2.5/gemlam')
myfiles = sorted(mypath.glob('gemlam_y2007*.nc'))
```

```python
for file in myfiles[100:]:
    orig = xr.open_dataset(file)
    newfile = (f'{file.name[0:6]}_3h{file.name[6:]}')
    print (file.name, newfile)
    three_hour = orig.resample({'time_counter': '3H'}).mean()
    three_hour.to_netcdf(newfile)
```

```python
orig = xr.open_dataset('/results/forcing/atmospheric/GEM2.5/gemlam/gemlam_y2007m06d15.nc')
threehourly = xr.open_dataset('gemlam_3h_y2007m06d15.nc')
```

```python
orig.qair[:, 250, 250].plot();
threehourly.qair[:, 250, 250].plot();
```

```python
orig.v_wind[:, 250, 250].plot();
threehourly.v_wind[:, 250, 250].plot();
```

```python
orig.u_wind[:, 250, 250].plot();
threehourly.u_wind[:, 250, 250].plot();
```

```python
plt.plot(orig.time_counter, np.sqrt(orig.u_wind[:, 250, 250]**2 + orig.v_wind[:, 250, 250]**2))
plt.plot(threehourly.time_counter, np.sqrt(threehourly.u_wind[:, 250, 250]**2 + threehourly.v_wind[:, 250, 250]**2));
```
