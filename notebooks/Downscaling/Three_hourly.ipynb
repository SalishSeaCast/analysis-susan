```python
# Method
file = '/results/forcing/atmospheric/GEM2.5/gemlam/gemlam_y2008m08d12.nc'
orig = xr.open_dataset(file)
start = dt.datetime(2008, 8, 12, 0)
endtime = dt.datetime(2008, 8, 12, 23)
timerange = arrow.Arrow.range('hour', start, endtime)
newtimes = []
for i, r in enumerate(timerange):
    newtimes.append(r.naive)

ds_new = orig.assign_coords(time_counter=newtimes)

ds_new.attrs['history'] = ds_new.attrs['history'] + f'{arrow.now()} Corrected time stamps'
ds_new.to_netcdf('TimeFixed/gemlam_y2008m08d12.nc')
```

```python
badfiles = ['gemlam_y2008m07d16.nc',
'gemlam_y2008m07d17.nc',
'gemlam_y2008m07d18.nc',
'gemlam_y2008m07d19.nc',
'gemlam_y2008m07d20.nc',
'gemlam_y2008m07d21.nc',
'gemlam_y2008m07d22.nc',
'gemlam_y2008m07d23.nc',
'gemlam_y2008m08d12.nc']

for file in badfiles:
    orig = xr.open_dataset(mypath / file)
    newfile = (f'{file[0:6]}_3h{file[6:]}')
    print (file, newfile)
#    print (int(file[8:12]), int(file[13:15]), int(file[16:18]))
    start = dt.datetime(int(file[8:12]), int(file[13:15]), int(file[16:18]), 0)
    endtime = dt.datetime(int(file[8:12]), int(file[13:15]), int(file[16:18]), 23)
    timerange = arrow.Arrow.range('hour', start, endtime)
    newtimes = []
    for i, r in enumerate(timerange):
        newtimes.append(r.naive)
#    print (newtimes)

    ds_new = orig.assign_coords(time_counter=newtimes)
    three_hour = ds_new.resample({'time_counter': '3H'}).mean()
    three_hour.to_netcdf(newfile)
```
