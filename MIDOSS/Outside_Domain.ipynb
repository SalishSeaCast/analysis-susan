```python
import matplotlib.pyplot as plt
import xarray as xr

%matplotlib inline
```

```python
bathy = xr.open_dataset('/home/sallen/MEOPAR/grid/bathymetry_201702.nc')
jjii = xr.open_dataset('/home/sallen/MEOPAR/grid/grid_from_lat_lon_mask999.nc')
```

```python
g88r64 = [-122.64, 47.142]
g88r70 = [-122.66, 47.133]
g94p101 = [-126.14, 50.475]
g94p102 = [-126.15, 50.472]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(9, 9))
ax.pcolormesh(bathy.nav_lon, bathy.nav_lat, bathy.Bathymetry);
for point in [g88r64, g88r70, g94p101, g94p102]:
    ax.plot(point[0], point[1], 'ro')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(18, 9))
for ax in axs:
    ax.pcolormesh(bathy.Bathymetry);
    for point in [g88r64, g88r70, g94p101, g94p102]:
        jj = jjii.jj.sel(lats=point[1], lons=point[0], method='nearest').item()
        ii = jjii.ii.sel(lats=point[1], lons=point[0], method='nearest').item()
        ax.plot(ii + 0.5, jj + 0.5, 'ro')
axs[0].set_ylim(880, 898)
axs[0].set_xlim(40, 60);
axs[1].set_ylim(0, 50)
axs[1].set_xlim(120, 170);
```
