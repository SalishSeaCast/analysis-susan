```python
watercolour = 'lightskyblue'
landcolour = 'papayawhip'
waterland_cmap = mplcolours.LinearSegmentedColormap.from_list('mycmap', [(0, watercolour),  (0.85, watercolour), 
                                                    (0.850001, landcolour), (1, landcolour)])
imin, imax = 220, 400
jmin, jmax = 120, 280
xs = range(jmin, jmax)
ys = range(imin, imax)
xx, yy = np.meshgrid(xs, ys)
```

```python
cnorm = mplcolours.Normalize(vmin=0, vmax=7)
mycmap = mpl.cm.copper
fig, axs = plt.subplots(1, 3, figsize=(15, 7))

for ax in axs:
    viz_tools.set_aspect(ax)
    ax.tick_params(axis='both', which='both', bottom=False, left=False, labelbottom=False, labelleft=False)
    ax.pcolormesh(1-mesh.tmask[0, 0, imin:imax, jmin:jmax], cmap=waterland_cmap)
    
fig.suptitle('10,000,000 $\ell$ Oil Spilled on Jan 15, 2018 at Salmon Bank');
ax_cbar = fig.add_axes([0.95, 0.12, 0.02, 0.76])
cb = colorbar.ColorbarBase(ax=ax_cbar, cmap=mycmap, norm=cnorm);
cb.set_label('Days to Beaching')

for iax, Lag in enumerate([Lagdiesel, LagAKNS, Lagbunker]):
    bt = np.array(Lag.Beaching_Time[imin:imax, jmin:jmax] - Lag.Beaching_Time[imin:imax, jmin:jmax].min())
    mymask = Lag.Beaching_Time[imin:imax, jmin:jmax] == Lag.Beaching_Time[imin:imax, jmin:jmax].min()
    days = np.ma.array(bt/1e9/86400., mask=mymask)
    axs[iax].scatter(xx-jmin+1, yy-imin+1, s=10, c=days, cmap='copper', vmin=0, vmax=7, marker='o', edgecolors='face');
    
axs[0].set_title('Diesel')
axs[1].set_title('AKNS')
axs[2].set_title('Bunker C');
```
